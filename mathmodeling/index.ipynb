

### 二階常微分方程示例
二階常微分方程  $ \frac{d^2x}{dt^2}=−Kx $（但是，  K>0 ）的示例。

這是一個單振動運動方程，可能是我成為大學生時遇到的第一個微分方程。

````python
x = Function('x')(t)
K = Symbol('K', positive = True)
eq = Eq(Derivative(x, t, 2), -K*x)
eq
````

````python
dsolve(eq, x)
````


如上所述，2個積分常數  C1,C2  顯示包含的一般解決方案。

作為初始條件，初始時刻  t=0  中的初始位置  x(0)=x0 ，

第一速度  ẋ(0)=v(0)=v0  作為解決這個運動方程...

````python
x0 = Symbol('x0')
v0 = Symbol('v0')

ans = dsolve(eq, x,
      ics = {x.subs(t, 0):x0,
             diff(x, t).subs(t, 0):v0}
      )
ans
````

## scipy.optimize 的 import
在圖表中，按月劃分的正常溫度看起來像 12 個月週期的正弦或餘弦函數。 現在，讓我們按如下方式擬合函數：

要適合函數，請執行以下操作： 嘗試使用以下函數進行擬合：scipy.optimizeimport

f(x)=β0+β1 cos(2πx/12)+β2sin(2πx/12)

````python
from scipy.optimize import leastsq
import numpy as np

def objectiveFunction(beta):
    r = y - theoreticalValue(beta)
    return r

def theoreticalValue(beta):  # sympy.cos ではなく，np.cos や np.pi を使う。
    f = beta[0] + beta[1] * np.cos(2*np.pi*x / 12) + beta[2] * np.sin(2*np.pi*x / 12)
    return f

x = np.array(month)
y = np.array(temp)

initialValue = np.array([10, 10, 1])
betaID = leastsq(objectiveFunction, initialValue)
betaID[0]
````

由最小二乘法確定  β0,β1,β2  值  f(x)  以繪製圖表。

````python
plt.plot(x, y, 'r.')
plt.plot(x, theoreticalValue(betaID[0]), 'b');
````

## 一、農夫一塊農地，打算種植小麥或大麥
假設一個農夫有一塊$A$平方千米的農地，打算種植小麥或大麥，或是兩者依某一比例混合種植。
該農夫只可以使用有限數量的肥料$F$ 和農藥$P$，而單位面積的小麥和大麥都需要不同數量的肥料和農藥，小麥以
$( F_{1},P_{1} )$表示，大麥以$ (F_{2},P_{2})$表示。
設小麥和大麥的售出價格分別為 $ S_{1} $ 和 $ S_{2} $，
則小麥與大麥的種植面積問題可以表示為以下的線性規劃問題：  
$ \max Z=S_{1}x_{1}+S_{2}x_{2}$(最大化利潤 - 目標函數)   
$ s.t.$   
$ x_{1}+x_{2}\leq A$  	(種植面積的限制)  
$ F_{1}x_{1}+F_{2}x_{2}\leq F$	(肥料數量的限制)   
$ P_{1}x_{1}+P_{2}x_{2}\leq P$ 	(農藥數量的限制)  
$ x_{1}\geq 0,\,x_{2}\geq 0$ 	（不可以栽種負數的面積）
