從上圖中看，兩個圖表似乎在一個點上相切。 實際上，我檢查它。

```python
# y = x**2 -1 と y = 4*x - 5 が交差または接している点の x 座標
def f(x):
    return x**2 - 1
def g(x):
    return 4*x - 5
# f(x) = g(x) をみたす x を求める。
sol = solve(Eq(f(x), g(x)), x)
sol
```

```python
# 解が1個なので接していることがわかった。その接点の座標は...
x0 = sol[0]
y0 = f(x0)
x0, y0
```

```python
# 接線の傾き
def df(x1):
    return diff(f(x), x).subs(x, x1)
df(x0)
```

```python
# 接線の方程式は
x, y = symbols('x y')
Eq(y, df(x0)*(x - x0) + y0)
```

### 海王星和冥王星的軌道
以焦點為原點的橢圓圖  
在太陽的萬有引力下移動的行星（不僅包括行星，而且包括準行星和小天體）繪製了以太陽為中心的橢圓軌道。 以焦點為原點，長半徑  a ， 偏轉率  e  橢圓的方程是極座標。  r,ϕ  可以表示為：  

$r=\frac{a(1−e2)}{1+e cos ϕ}$
 
冥王星曾經是第九顆行星，現在也是準行星之一，也繪製了橢圓軌道。 冥王星的軌道長半徑  aP=39.767au  ， 偏轉率  eP=0.254  繪製冥王星的軌道。

首先，將極座標顯示橢圓的表達式定義為函數。



```python
def r(a, e, phi):
    return a*(1-e**2)/(1 + e*cos(phi))
```

```python
import matplotlib.pyplot as plt
# デフォルトでは plt.rcParams['figure.figsize'] = (6.0, 4.0) だった
plt.rcParams['figure.figsize'] = (6, 6)
aP = 39.767
eP = 0.254
plot_parametric(r(aP,eP,phi)*cos(phi), r(aP,eP,phi)*sin(phi),
                (phi, 0, 2*pi), xlim=(-50,50), ylim=(-50,50));
```

```python
plt.plot(x, y, 'r.')
plt.plot(x, theoreticalValue(betaID[0]), 'b');
```

更平滑地繪製擬合函數。

```python
beta0, beta1, beta2 = betaID[0]

def ftemp(month):
    r = beta0 + beta1 * np.cos(2*np.pi*month/12) + beta2 * np.sin(2*np.pi*month/12)
    return r

month1 = np.arange(1, 12.1, 0.1)
temp1 = ftemp(month1)

plt.plot(x, y, 'r.')
plt.plot(month1, temp1, 'b');
```
